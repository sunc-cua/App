```python
import pandas as pd
import numpy as np
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI

data_dict = {
	"country": ["New York City", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"],
	"fake pop":        [500_000,     250_000,    800_000,    350_000,   600_000,        450_000,     200_000,      0,         456_123,    np.nan ],
	"happiness":       [9.94,           7.16,       6.35,       8.07,      6.98,           6.1,       4.23,       8.22,         6.87,      np.nan],
}

df = pd.DataFrame(data_dict)
df.head()

```

```python
# for LLM prompt
llm = OpenAI(api_token="") # deleted intentionaly.
pandas_ai = PandasAI(llm, conversational=False)
```

```python
# finding index of a row using value of a column
question = input("What is your question? ")
response = pandas_ai(df, question)
print(response)
```
